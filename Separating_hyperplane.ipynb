```python
def cross(a,b):
    return a[0]*b[1]-a[1]*b[0]

def sub(x, y):
    return list(map(lambda a, b: a - b, x, y))

def intersect(p1,p2,p3,p4):
    x1_ = min(p1[0],p2[0])
    y1_ = min(p1[1],p2[1])
    x2_ = max(p1[0],p2[0])
    y2_ = max(p1[1],p2[1])
    
    x3_ = min(p3[0],p4[0])
    y3_ = min(p3[1],p4[1])
    x4_ = max(p3[0],p4[0])
    y4_ = max(p3[1],p4[1])
    if (x2_ >= x3_) and (x4_ >= x1_) and (y2_ >= y3_) and (y4_ >= y1_):
        if cross(sub(p3,p1),sub(p2,p1))*cross(sub(p4,p1),sub(p2,p1))<= 0:
            if cross(sub(p1,p3),sub(p4,p3))*cross(sub(p2,p3),sub(p4,p3))<=0:
                return True
    return False

def rotate(A,B,C):
    return (B[0]-A[0])*(C[1]-B[1])-(B[1]-A[1])*(C[0]-B[0])

def jvs(A):
    n = len(A)
    P = [i for i in range(n)]
    for i in range(1,n):
        if A[P[i]][0]<A[P[0]][0]: 
            P[i], P[0] = P[0], P[i]  
    H = [P[0]]
    del P[0]
    P.append(H[0])
    while True:
        right = 0
        for i in range(1,len(P)):
            if rotate(A[H[-1]],A[P[right]],A[P[i]])<0:
                right = i
        if P[right]==H[0]: 
            break
        else:
            H.append(P[right])
        del P[right]
    return H   

def in_(p1,p2,p3):
    if cross(sub(p1,p2),sub(p3,p1)) <= 0:
        return True
    return False

def checkEqual(array):
    return len(set(array)) <= 1

def cross_line(a,b,c):
    dxc = c[0] - a[0]
    dyc = c[1] - a[1]
    
    dxl = b[0] - a[0]
    dyl = b[1] - a[1]

    return dxc * dyl - dyc * dxl

def check(dark,light):
    if len(dark) < len(light):
        temp = list(dark)
        dark = list(light)
        light = list(temp)
    d = jvs(dark)
    l = jvs(light)
    l_len = len(l)
    d_len = len(d)
    in_pd = [_ for _ in range(d_len)]
    in_pl = [_ for _ in range(l_len)]
    if (l_len>1):
        for i in range(d_len):
            for k in range(l_len):
                #print(dark[d[i]],dark[d[(i+1)%d_len]],light[l[k]],light[l[(k+1)%l_len]])
                if intersect(dark[d[i]],dark[d[(i+1)%d_len]],light[l[k]],light[l[(k+1)%l_len]]):
                    print("No")
                    return
        for k in l:
            for i in range(d_len):
                in_pd[i] = in_(dark[d[i]],dark[d[(i+1)%d_len]],light[k])
            for di in d:
                if cross_line(dark[d[0]],dark[d[d_len-1]],dark[di]) != 0:
                    if checkEqual(in_pd):
                        print("No")
                        return
            in_pd = [_ for _ in range(d_len)]
        for k in d:
            for i in range(l_len):
                in_pl[i] = in_(light[l[i]],light[l[(i+1)%l_len]],dark[k])
            for li in l:
                if cross_line(light[l[0]],light[l[l_len-1]],light[li]) != 0:
                    if checkEqual(in_pl):
                        print("No")
                        return
            in_pl = [_ for _ in range(l_len)]        
    else:
        for i in range(d_len):
            if intersect(dark[d[i]],dark[d[(i+1)%d_len]],light[0],light[0]):
                print("No")
                return
            in_pd[i] = in_(dark[d[i]],dark[d[(i+1)%d_len]],light[0])
        for di in d:
            if cross_line(dark[d[0]],dark[d[d_len-1]],dark[di]) != 0:
                if checkEqual(in_pd):  
                    print("No")
                    return      
    print("Yes")
    return

if __name__ == "__main__":
    N = int(input())
    dark = [[] for _ in range(1)]
    light = [[] for _ in range(1)]
    for i in range(N):
        temp = [int(i) for i in input().split()]
        t = temp[:2]
        if temp[2]==1:
            dark.append(t)
        else:
            light.append(t)
    dark.pop(0)
    light.pop(0)
    if (N==1) or not dark or not light:
        print("Yes")
    else:
        check(dark,light)
```
